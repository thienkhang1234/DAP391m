```python
df = pd.read_csv("/content/titanic.csv")

df['sex'] = df['sex'].replace({1: 'Female', 0: 'Male'}).astype('category')


df['embarked'] = df['embarked'].replace({
    'C': 'Cherbourg',
    'Q': 'Queenstown',
    'S': 'Southampton'
}).astype('category')


df['survived'] = df['survived'].replace({1: 'Yes', 0: 'No'}).astype('category')


pclass_order = ['1st', '2nd', '3rd']
df['pclass'] = df['pclass'].replace({1: '1st', 2: '2nd', 3: '3rd'})
df['pclass'] = pd.Categorical(df['pclass'], categories=pclass_order, ordered=True)


print(df[['sex', 'embarked', 'survived', 'pclass']].head())
print("\nMissing values check:")
print(df[['sex', 'embarked', 'survived', 'pclass']].isna().sum())

df.head()
```
